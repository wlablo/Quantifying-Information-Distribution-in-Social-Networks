```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
import matplotlib.cm as cm

# === Load dataset ===
df = pd.read_csv("TwiBot-#Zandberg-20250308_174911.csv", dtype=str)

def norm_handle(s):
    if pd.isna(s): return None
    s = str(s).strip()
    if not s.startswith("@"): s = "@" + s
    return s

df["handle"] = df["handle"].apply(norm_handle)

def parse_mentions(s):
    if pd.isna(s): return []
    out = []
    for m in str(s).split(","):
        m = m.strip()
        if not m: continue
        if not m.startswith("@"): m = "@" + m
        out.append(m)
    return out

df["mentions_parsed"] = df["mentions"].apply(parse_mentions)

# === Build weighted directed graph ===
G = nx.DiGraph()
for _, row in df.iterrows():
    src = row["handle"]
    if not src: continue
    for dst in row["mentions_parsed"]:
        if src == dst: continue
        if G.has_edge(src, dst):
            G[src][dst]["weight"] += 1
        else:
            G.add_edge(src, dst, weight=1)

UG = G.to_undirected()

# === Centralities (normalized to [0,1]) ===
def normalize(d):
    m = max(d.values())
    return {k: v/m for k,v in d.items()}

deg = nx.degree_centrality(UG)
bet = nx.betweenness_centrality(UG, normalized=True)
clo = nx.closeness_centrality(UG)

deg_n, bet_n, clo_n = map(normalize, [deg, bet, clo])

# === Function to draw subtle colored graphs ===
def draw_colored_graph(G, centrality, title, filename):
    plt.figure(figsize=(6,6))
    pos = nx.circular_layout(G)
    vals = list(centrality.values())
    cmap = cm.viridis
    colors = [cmap(v*0.9 + 0.1) for v in vals]
    sizes = [300 * (v + 0.05) for v in vals]
    nx.draw_networkx_edges(G, pos, edge_color="gray", width=0.4, alpha=0.4)
    nx.draw_networkx_nodes(G, pos, node_size=sizes, node_color=colors, alpha=0.9, linewidths=0)
    plt.title(title, fontsize=12)
    plt.axis("off")
    plt.tight_layout()
    plt.savefig(filename, dpi=150, bbox_inches="tight", facecolor="white")
    plt.close()

draw_colored_graph(UG, deg_n, "Weighted Degree Centrality (normalized)", "fig_network_degree_circular.png")
draw_colored_graph(UG, bet_n, "Betweenness Centrality (normalized)", "fig_network_betweenness_circular.png")
draw_colored_graph(UG, clo_n, "Closeness Centrality (normalized)", "fig_network_closeness_circular.png")

print("✅ Saved 3 circular graphs with subtle color gradients.")
```

```python
# === 1. Importy ===
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
import matplotlib.cm as cm

# === 2. Wczytaj dane ===
df = pd.read_csv("TwiBot-#Zandberg-20250308_174911.csv", dtype=str)

# Normalizacja nazw użytkowników
def norm_handle(s):
    if pd.isna(s): 
        return None
    s = str(s).strip()
    if not s.startswith("@"): 
        s = "@" + s
    return s

df["handle"] = df["handle"].apply(norm_handle)

# === 3. Parsowanie kolumny mentions ===
def parse_mentions(s):
    if pd.isna(s): 
        return []
    out = []
    for m in str(s).split(","):
        m = m.strip()
        if not m: 
            continue
        if not m.startswith("@"): 
            m = "@" + m
        out.append(m)
    return out

df["mentions_parsed"] = df["mentions"].apply(parse_mentions)

# === 4. Budowa grafu ważonego ===
G = nx.DiGraph()
for _, row in df.iterrows():
    src = row["handle"]
    if not src:
        continue
    for dst in row["mentions_parsed"]:
        if src == dst:
            continue
        if G.has_edge(src, dst):
            G[src][dst]["weight"] += 1
        else:
            G.add_edge(src, dst, weight=1)

UG = G.to_undirected()

print(f"✅ Graph built: {G.number_of_nodes()} nodes, {G.number_of_edges()} edges")

# === 5. Centralności ===
deg = nx.degree_centrality(UG)
bet = nx.betweenness_centrality(UG, normalized=True)
clo = nx.closeness_centrality(UG)

# Normalizacja centralności
def normalize(d):
    m = max(d.values())
    return {k: v / m for k, v in d.items()}

deg_n, bet_n, clo_n = map(normalize, [deg, bet, clo])

# === 6. Funkcja rysująca graf circular ===
def draw_colored_graph(G, centrality, title, filename):
    # Utworzenie figury i osi
    fig, ax = plt.subplots(figsize=(6, 6))
    
    # Circular layout
    pos = nx.circular_layout(G)
    vals = np.array(list(centrality.values()))
    cmap = cm.viridis
    colors = [cmap(v*0.9 + 0.1) for v in vals]
    sizes = [300 * (v + 0.05) for v in vals]

    # Krawędzie i węzły
    nx.draw_networkx_edges(G, pos, ax=ax, edge_color="gray", width=0.4, alpha=0.4)
    nx.draw_networkx_nodes(G, pos, ax=ax, node_size=sizes, node_color=colors, alpha=0.9, linewidths=0)

    # Utworzenie legendy (colorbar) powiązanej z osią ax
    sm = plt.cm.ScalarMappable(cmap=cmap, norm=plt.Normalize(vmin=0, vmax=1))
    sm.set_array([])
    cbar = fig.colorbar(sm, ax=ax, shrink=0.75, pad=0.02)  # <-- kluczowe: fig.colorbar zamiast plt.colorbar
    cbar.set_label("Normalized Centrality", fontsize=10)
    cbar.ax.tick_params(labelsize=8)

    ax.set_title(title, fontsize=12)
    ax.axis("off")
    plt.tight_layout()
    plt.savefig(filename, dpi=150, bbox_inches="tight", facecolor="white")
    plt.show()

# === 7. Generowanie wykresów ===
draw_colored_graph(UG, deg_n, "Weighted Degree Centrality (normalized)", "fig_network_degree_circular.png")
draw_colored_graph(UG, bet_n, "Betweenness Centrality (normalized)", "fig_network_betweenness_circular.png")
draw_colored_graph(UG, clo_n, "Closeness Centrality (normalized)", "fig_network_closeness_circular.png")

print("✅ All 3 circular graphs generated and saved.")
```

```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np

# === Funkcja rysująca histogram z etykietami i kolorami ===
def plot_histogram(data, title, filename, bins=30):
    vals = np.array(list(data.values()))
    cmap = cm.viridis
    color = cmap(0.65)

    plt.figure(figsize=(7,4))
    plt.hist(vals, bins=bins, color=color, edgecolor='black', alpha=0.8)
    plt.xlabel("Normalized Centrality", fontsize=11)
    plt.ylabel("Frequency", fontsize=11)
    plt.title(title, fontsize=13)
    plt.grid(alpha=0.3, linestyle='--', linewidth=0.5)
    plt.tight_layout()
    plt.savefig(filename, dpi=150, bbox_inches="tight", facecolor="white")
    plt.show()

# === Rysowanie histogramów ===
plot_histogram(deg_n, "Distribution of Normalized Degree Centrality", "hist_degree.png")
plot_histogram(bet_n, "Distribution of Normalized Betweenness Centrality", "hist_betweenness.png")
plot_histogram(clo_n, "Distribution of Normalized Closeness Centrality", "hist_closeness.png")

print("✅ All histograms generated and saved.")
```
